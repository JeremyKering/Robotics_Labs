```python
def hom_transform(theta, x, y, z):
    """
    Creates a 4x4 homogeneous transformation matrix
    with a rotation about the Z-axis and translation (x, y, z).

    Parameters:
        theta : float
            Rotation angle in degrees
        x, y, z : float
            Translation components
    Returns:
        T : numpy.ndarray
            4x4 homogeneous transformation matrix
    """
    theta = math.radians(theta)
    # Rotation matrix about Z-axis
    R = np.array([
        [np.cos(theta), -np.sin(theta), 0],
        [np.sin(theta),  np.cos(theta), 0],
        [0,              0,             1]
    ])

    # Translation vector
    p = np.array([x, y, z])

    # Homogeneous transformation matrix
    T = np.eye(4)
    T[:3, :3] = R
    T[:3, 3] = p

    return T
```

$$
R_1 = \begin{bmatrix}

$$
T_1 = \begin{bmatrix}

\xi_3 = \xi_1 \times \xi_2 = T_1 \times T_2
